```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3, 3001.76]

# Create a line chart, years on x axis and  gdp on y axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# Add a label to the y-axis:
plt.ylabel("Billions of $")
plt.show()
```

```python
movies = ["A", "B", "C", "D", "E", "F"]
num_oscars = [5,11,3,8,10,7]

# increase bar size by 0.1 from 0.8:
# enumerate: use both a lists index and value
xs = [i+0.1 for i, _ in enumerate(movies)]

# plot bars with x-coords on left [xs], and heights on y-axis [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Awards")
plt.title("Movie Oscar Count")

# Label x-axis with movie names at bar centers
plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)

plt.show()
```

```python
# Line Chart good for trends:
variance = [1,2,4,8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance, bias_squared)]

xs = [i for i, _ in enumerate(variance)]
#call plot

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label = 'total error')

# loc=9: top center
plt.legend(loc=9)
plt.xlabel("Model Complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
# Scatterplot: Visualize the relationship between 2 paired sets of data
# Check number of friends on socail media platform against time spent on site
friends = [70, 110,34, 29, 123, 943, 355, 762, 10]
minutes = [30, 50, 12, 56, 89, 290, 23, 45, 18]
labels = ['a', 'b','c','d','e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

#label each point:
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, # assign label
                xy=(friend_count, minute_count), # graph the point
                xytext=(5,-5), #offset text
                textcoords='offset points')
    
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("# of minutes")
plt.show()
```
